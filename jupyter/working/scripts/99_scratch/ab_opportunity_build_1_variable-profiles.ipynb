```python
tr = Transition('opportunities-many')
db = DataBuilder("opportunities-many")
tools = db.tools
```

```python
tr_accounts = Transition('accounts-many')
df = tr_accounts.get_source_data()
sample_size = df.shape[0]
```

```python
df_names = tools.get_profiles(size = 10000, mf_weighting=[7,3], seed = 1)

associate = [{'gender': {'expect': 'category','value': ['M', 'F']}}]
actions = {0: {'action': 'get_custom', 'kwargs' : {'code_str': "f'{first_name} {last_name}'", 'first_name': {'_header': 'forename'},'last_name': {'_header': 'surname'}}}}
df_names['SponsorName']= tools.associate_dataset(df_names, associations=associate, actions=actions)
```

```python
tr.discover.data_dictionary(df_names)
```

```python
df_names = pd.DataFrame(np.repeat(df_names.values,100,axis=0), columns=df_names.columns)
```

```python
sponsor_selection = df_names['SponsorName']
sponsor_weighting = [2,5,3,4,3,4,7,8.4,10]
df['Sponsor'] = tools.get_category(selection = sponsor_selection, weight_pattern=sponsor_weighting, size = sample_size)
```

```python
df['Opportunity_Id'] = tools.get_string_pattern('006dUdddddlduddlUlU',choice_only=False, size = sample_size, seed = 1)
```

```python
opportunity_selections = ["Engage - QuickStart(Customer360)" , "Engage - Enterprise(Conversational AI)", "Engage - Enterprise(Customer360)" , "Engage - QuickStart(Coversational AI)","Amplify - QuickStart(Process Insights)" , "Amplify - Enterprise(Process Insights)" , "Amplify - QuickStart(Survellance Insights)", "Amplify - Enterprise(Survellance Insights)", "Cortex QuickStart(MachineLearning)", "Cortex QuickStart(AI Develpoment)","Cortex Enterprise(MachineLearning)", "Cortex Enterprise(AI Develpoment)"] 
opportunity_weighting = [4,1,1,3,3,1,1,4,3,3,1,2]
df['Opportunity_Name'] = tools.get_category(selection = opportunity_selections, weight_pattern=opportunity_weighting, size = sample_size, seed = 1)
```

```python
status_selection = ['Proposed Solution','Prospecting','Confirming Value with Power','Use Case Confirmed', 'Needs Identification', 'Negotiations']
status_weighting = [2,5,1,3,3,4]
df['Status'] = tools.get_category(selection = status_selection, weight_pattern=status_weighting, size = sample_size, seed = 1)
```

```python
date_weighting = list(np.exp(np.arange(-5, 0.0, 0.2)).round(2))+[1,2,5,10,11,10,9,8,5]
until = datetime.datetime.today()
start = until - pd.DateOffset(years = 2)

df['LastModifiedDate'] = tools.get_datetime(start= start, until= until, date_pattern=date_weighting, date_format= '%Y-%m-%d %H:%M:%S' , size=sample_size, seed = 1)
```

```python
opportunity_correlation = [["Engage - QuickStart(Customer360)" ,"Engage - QuickStart(Coversational AI)","Amplify - QuickStart(Process Insights)", "Amplify - QuickStart(Survellance Insights)",  "Cortex QuickStart(MachineLearning)", "Cortex QuickStart(AI Develpoment)"],[ "Engage - Enterprise(Coversational AI)", "Engage - Enterprise(Customer360)" , "Amplify - Enterprise(Process Insights)", "Amplify - Enterprise(Process Insights)", "Amplify - Enterprise(Survellance Insights)","Cortex Enterprise(MachineLearning)", "Cortex Enterprise(AI Develpoment)" ]] 
opportunity_action = {0: {'action': 50000}, 
                      1: {'action': 'get_number', 'kwargs' : {'from_value':100, 'to_value' : 50000, 'weight_pattern': [1,3,5,4,3,1]+[1]*5, 'offset': 1000}}}
df['Amount'] = tools.correlate_categories(df['Opportunity_Name'], correlations = opportunity_correlation, actions= opportunity_action, value_type= "category" ) 
```

```python
df2 = pd.read_csv("../../data/0_raw/contacts.csv")
df2 = df2.rename(index=str, columns={"Company Name": "AccountName", "Employee Title" : "ContactTitle"})
df2["ContactName"] = df2["Employee First Name"]+ " " +df2["Employee Last Name"]
df2 = df2.drop(columns = ['Employee First Name', 'Employee Last Name'])
df2 = df2.drop_duplicates(["AccountName"])
```

```python
df = pd.merge(df,df2, on='AccountName', how= 'left')
```

```python
analysis = discover.analyse_category(df['ContactName'])
size = len(df['ContactName'][df['ContactName'].isna()])
df_names = tools.get_profiles(size = size, mf_weighting=[6,4], seed = 1)
associate = [{'gender': {'expect': 'category','value': ['M', 'F']}}]
actions = {0: {'action': 'get_custom', 'kwargs' : {'code_str': "f'{first_name} {last_name}'", 'first_name': {'_header': 'forename'},'last_name': {'_header': 'surname'}}}}
df_names= tools.associate_dataset(df_names, associations=associate, actions=actions)
df['ContactName'][df['ContactName'].isna()] = df_names
```

```python
df['ContactEmail'] = df["ContactName"].str.lower().str.replace(' ', '')+ "@"+ df['AccountName'].str.lower().str.replace(' ', '')+ ".com"
```
